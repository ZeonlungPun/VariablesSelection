```python
import keras
import tensorflow as tf
import vis
import matplotlib.pyplot as plt
import numpy as np
```

```python
from keras.applications.vgg16 import VGG16,preprocess_input
model=VGG16(weights='imagenet')
```

```python
import json
class_index=json.load(open("imagenet_class_index.json"))
classlabel=[]
for i_dict in range(len(class_index)):
  classlabel.append(class_index[str(i_dict)][1])
```

```python
from keras.layers import preprocessing
from keras.utils.image_utils import load_img,img_to_array

img=load_img('190812_EN_陳曉華1.jpg',target_size=(224,224))
img=img_to_array(img)
img=preprocess_input(img)
y_pred=model.predict(np.expand_dims(img,0))
idxs=np.argsort(y_pred.flatten())[::-1]
for i,idx in enumerate(idxs[:5]):
  print("top{}class: name={},probability={}".format(i+1,classlabel[idx],y_pred[0,idx]))
```

```python
from vis.utils import utils
layer_idx=utils.find_layer_idx(model,'predictions')
#swap softmax with linear
model.layers[layer_idx].activation=keras.activations.linear
model=utils.apply_modifications(model)
```
